```python
print("Representing the Graphs and Plots:")
```

```python
x = data["Attendance (%)"]
y = data["Final Exam Marks (out of 100)"]

plt.plot(x, y, 'o') 
plt.title("Attendance vs Final Exam Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Final Exam Marks")
plt.show()
```

```python
x = data["Attendance (%)"]
y = data["Daily Study Hours"]

plt.plot(x, y, 'o', color="red") 
plt.title("Attendance vs Daily Study Hours")
plt.xlabel("Attendance (%)")
plt.ylabel("Daily Study Hours")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
attendance = data["Attendance (%)"]
study_hours = data["Daily Study Hours"]
final_marks = data["Final Exam Marks (out of 100)"]

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.plot(attendance, final_marks, 'o', color="blue")
plt.title("Attendance vs Final Exam Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Final Exam Marks")

plt.subplot(2,3,2)
plt.plot(study_hours, final_marks, 'o', color="green")
plt.title("Study Hours vs Final Exam Marks")
plt.xlabel("Daily Study Hours")
plt.ylabel("Final Exam Marks")

plt.subplot(2,3,3)
plt.hist(attendance, bins=15, color="skyblue", edgecolor="black")
plt.title("Distribution of Attendance")
plt.xlabel("Attendance (%)")
plt.ylabel("Count")

plt.subplot(2,3,4)
plt.hist(study_hours, bins=10, color="orange", edgecolor="black")
plt.title("Distribution of Study Hours")
plt.xlabel("Daily Study Hours")
plt.ylabel("Count")

plt.subplot(2,3,5)
plt.hist(final_marks, bins=15, color="purple", edgecolor="black")
plt.title("Distribution of Final Exam Marks")
plt.xlabel("Final Exam Marks")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```
